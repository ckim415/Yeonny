```python
vowels = ('a', 'e', 'i', 'o', 'u')
l,c = map(int, input().split())
for pw in combinations(sorted(letters), 4):
  
  count = 0
  for i in pw:
    if i in vowels:
      count += 1

  if count >= 1 and count <= l-2:
    print(pw)
```

```python
import copy
```

```python
result = []
result.append(copy.deepcopy('a'))
result.append(copy.deepcopy('abc'))
```

```python
result = []
string = []
visited = []

def combination(letters, length, index):
  if len(string) == length:
    result.append(copy.deepcopy(string))

  for i in range(index, len(letters)):
    if i in visited:
      continue

    string.append(letters[i])
    visited.append(i)
    print('before',index, string, visited, result)
    combination(letters, length, index+1)
    print('after',index, string, visited, result)
    string.pop()
    visited.pop()
    

    
```

```python
combination(letters, 4, 0)
```
